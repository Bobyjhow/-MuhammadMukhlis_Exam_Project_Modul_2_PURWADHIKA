# Latar Belakang <br>

Youtube adalah website tempat berbagi video secara gratis. Walaupun pengguna Youtube mengakses secara gratis, Youtube tetap membayar sejumlah uang yang besarnya bervariasi kepada pembuat atau penyedia video (*content creator*) di dalamnya. Bayaran yang menggiurkan dan tidak adanya batasan usia untuk menjadi *content creator*, membuat Youtube dijadikan sebagai pekerjaan tetap bagi para *content creator* . Selama *content creator* dapat menyediakan video yang dapat ditonton oleh banyak orang, Youtube tidak segan-segan akan membayar sejumlah uang tergantung dari banyaknya *view* pada video yang dibuat. Namun akun atau *channel* milik *content creator* harus diverifikasi terlebih dahulu sehingga akan mengikat pada term and agreement. Dilihat dari cara mendapatkan uang pada Youtube yang tergolong mudah, tak heran banyak orang yang berbondong-bondong untuk menjadi *content creator* pada *website* raksasa ini. Namun ada hal yang perlu diperhatikan bagi para *content creator* pemula yaitu bagaimana mereka "menjual" video mereka supaya dapat di-*view* oleh banyak orang sehingga *channel* mereka dapat berkembang dan dikenal. Youtube memiliki fitur berlangganan bernama "*subscribe*", fitur ini menjadi daya tarik sebuah *channel* dengan menampilkan berapa banyak *subscriber* yang ada pada *channel* tersebut sehingga banyak menarik perhatian banyak orang. Bagi pemula, menaikkan jumlah *subscriber* bukan perkara yang mudah. Mereka harus bekerja keras untuk mengiklankan video mereka kepada orang-orang. Namun ada salah satu cara supaya orang-orang dapat melirik *channel* para pemula dengan memanfaatkan fitur "*Youtube Trending Videos*". Fitur ini akan menjajakan video apa saja yang sedang tren di beranda pengguna Youtube. Tren video dikategorikan baik dari tiap negara sampai seluruh dunia tak terkecuali Negara US. Tidak peduli video yang sedang tren berasal dari *channel* yang terkenal ataupun tidak, selama video ditonton oleh banyak orang, dapat dikatakan "pasti" orang-orang yang menonton video tersebut akan mengunjungi *channel* dari pemilik video tren tersebut. Hal ini menjadi kesempatan besar bagi para pemula supaya *channel*-nya dilirik oleh banyak orang yang berpotensi menjadi *subscriber*.

Kita melihat data outlier sangat mengganggu perhitungan rata-rata (nilai datanya sampai ribuan). Maka dari itu mari kita hapus seluruh data *outlier* dari kolom `days_until_trending`. Mengapa dihapus? karena mengingat tidak ada batasan waktu dari video tayang sampai menjadi tren tentu tidak akan menjadi masalah. Ditambah jika kita ubah data *outlier* menjadi median yaitu 6 (kita bandingkan dengan median karena persebaran data selalu di dekat median), maka data akan condong ke nilai 6 (data *outlier* berjumlah 1249). Sehingga lebih baik kita menghapus data *outlier*-nya.

```python

print("Nilai median dari data:",df['days_until_trending'].median())
```

```python
Q1 = df['days_until_trending'].quantile(0.25)
Q3 = df['days_until_trending'].quantile(0.75)
IQR = Q3 - Q1

lower_bound =  Q1 - (1.5 * IQR)
upper_bound =  Q3 + (1.5 * IQR)

print("Banyaknya data Outlier:",df[df['days_until_trending']> upper_bound]['days_until_trending'].count()) #banyak data outlier
```

```python
listt = []
for i in df['days_until_trending'] :
    if lower_bound < i < upper_bound:
        listt.append(i)
        
daysuntiltrending = pd.DataFrame(data = listt)
daysuntiltrending.rename(columns = {0:'Days Until Trending'}, inplace = True)
display(daysuntiltrending.head(), daysuntiltrending.tail()) # hasilnya setelah data outlier dari kolom days_until_trending dihapus
```

Sekarang kita coba uji korelasi antara kombinasi dari `days until trending`, `views`, `likes`, `dislikes`, dan `comment count`.

Sebelum kita menguji korelasi, kita harus pastikan apakah data berdistribusi normall atau tidak. Untuk menguji normalitas dari suatu data terdapat banyak metode utnuk melakukan pengujiannya. Pengujian data ini dilakukan dengan metode `Shapiro Wilk` Test. Dari yang kita pelajari, metode `Shapiro Wilk` adalah salah satu metode terbaik untuk menguji normalitas.

```python
'''  
*) H0: Data berdistribusi Normal
*) Ha: Data tidak berdistribusi normal
'''
a = 0.05
stat, pval = shapiro(df[['days_until_trending', 'views', 'likes', 'dislikes', 'comment_count']]) # uji normalitas shapiro


if pval < a:
    print('Tolak H0')
else:
    print('Terima H0')
```

Nilai P-Value kurang dari 0.05 maka kita menolak H0 dan menyatakan bahwa data tidak terdistribusi dengan normal. Karena data tidak berdistribusi normal atau non-parametrik, kita gunakan metode spearman untuk menguji korelasi.

```python
corr_matrix = df[['days_until_trending','views', 'likes', 'dislikes', 'comment_count']].corr(method= 'spearman')
print(corr_matrix)
plt.figure(figsize = (12,10))
plt.title("Correlation Matrix")
sns.heatmap(corr_matrix, annot=True)
plt.show()
```

Dari data di atas terlihat bahwa data kolom `views` adalah faktor utama video menjadi trending. Hal ini senada dari pernyataan resmi youtube yang mengatakan *view* adalah faktor utama video menjadi trending. Ditambah, video akan dibandingkan performanya dengan video lainnya pada *channel* yang sama. Artinya selama *view* dari video bisa meningkat pesat dari video lainnya dari *channel* yang sama, maka kesempatan video tersebut menjadi *trending* terbuka lebar.
